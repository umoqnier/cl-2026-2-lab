```python
# Instalación de librerías externas
# torch: framework de deep learning para construir la red neuronal
# subword-nmt: implementación del algoritmo BPE (Byte-Pair Encoding)
# rich: impresión con colores en consola (más legible)
!pip install -q torch subword-nmt rich
```

```python
# Descarga de corpus y modelos de NLTK
import nltk

recursos = [
    "gutenberg",   # Clásicos de la literatura
    "abc",         # Noticias australianas
    "inaugural",   # Discursos inaugurales de presidentes de EE.UU.
    "state_union", # Discursos del Estado de la Unión
    "webtext",     # Textos web variados
    "genesis",     # Corpus de génesis (¡este es nuestro conjunto de prueba!)
    "punkt_tab",   # Tokenizador de oraciones
]

for recurso in recursos:
    nltk.download(recurso, quiet=True)

print("✅ Recursos de NLTK descargados correctamente")
```
